
#**ETL vs ELT processes**

```python
##ETL: definition, pros and cons, tools, business use cases 

#Definition
Data integration process that helps organizations extracting data from various sources and bring it into a single target database.
It has 3 steps:
- Extraction: data is extracted from source systems using DB queries or change data capture processes
- Following the extraction, data is moved to a staging area
- Transformation: data is cleaned, processed and turned into a common format 
    so it can be consumed by a target data warehouse for analysis  by a BI tool (Tableau, Looker, PowerBI)
- Loading: formated data is loaded into target system (object type, creating schemas) 

# Pros
- Preserves Resources - can reduce the volume of data that is stored in the warehouse and preserve storage bandwith and computation resources
- Improves compliance/Granular compliance on in-flight data - can mask and remove sensitive data, encrypting specific informations that helps companies comply with data privacy and protection
- Tranformations with business logic
- Well developed tools : ETL has existed for decades and there is a range of robust platforms that business can deploy in this approach

#Cons 
- Legacy ETL is slow (required disk based staging and transformations)
- Frequent maintenance
- Higher Upfront Cost: defining business logic can increase the scope of a data integration project 
    
#Modern streaming ETL platforms: Apache Kafka, Spark Streaming
- Deliver real time data integration (in memory stream processing technology)
- Data is loaded in real time 
- while transformation logic is compiled and processed in memory (faster than disk based processing)
- and scaled across multiple nodes to handle high volumes of data at sub second speeds 
```

```python
##ELT: definition, pros and cons, tools, business use cases 

# Definition
ELT is a data integration process that transfers data from a source system into a target system without business logic-driven transformations on the data. 
The ELT process involves three stages:
- Extraction: raw data is extracted from various sources
- Loading: data is delivered directly to target system
- Tranformation: the target platform can transform the data for report purposes 

#What is the difference from ELT?
An ELT pipeline reorders the steps involved in the integration process with the data transformation step 
occurring at the end instead of in the middle of the process.

#Why did it emerged?
Cloud storage and analytics resources became more affordable and powerful
--> ETL pipelines became ill suited to deal with growing variety and volume of data created in the cloud based services
--> Companies can now afford to store and process all their unstructured data in the cloud (no need to reduce or filter data during the transformation)

    
#Pros:
- Fast extraction and loading
- More flexibility
- Lower upfront development costs 
- Low latency in streaming 
- Rapid data loading
- Minimal maintenance
- Highly automated workflow

#Cons 
- Overgeneralization
- Security gaps
- Compliance risk (privacy regulations)
- Increased latency in cases where transformations with business logic are needed 




```
