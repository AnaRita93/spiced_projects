```python
#Latent Dirichlet Allocation (LDA) Topic Modeling
#Sources: https://www.tidytextmining.com/topicmodeling.html#library-heist

# After the data is vectorized, we can use LDA approach to find out topics present in the corpus of the fited model

lda_model = LatentDirichletAllocation(n_components=10, learning_method='online', max_iter=500, random_state=0).fit(X)

# Lets find out the Top 10 topics and display them using a function
# Each topic will have 5 words 

no_top_words = 5

#the function will have 3 parameters, LDA model, feature names from the term matrix and the number of words per topic 
def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic %d:" % (topic_idx))
        print(" ".join([feature_names[i]
                          for i in topic.argsort()[:-no_top_words - 1:-1]]))
```

```python

display_topics(lda_model, vectorizer.get_feature_names(), no_top_words)
```

```python
#Term Frequency Analysis 

import nltk
nltk.download('punkt')


lyrics_train['tokenized'] = lyrics_train['lyrics'].apply(word_tokenize)
lyrics_train.head()
```

```python
lyrics_train['lower'] = lyrics_train['tokenized'].apply(lambda x: [word.lower() for word in x])
lyrics_train.head()
```

```python
punc = string.punctuation
lyrics_train['no_punc'] = lyrics_train['lower'].apply(lambda x: [word for word in x if word not in punc])
lyrics_train.head()
```

```python
stop_words = set(stopwords.words('english'))
lyrics_train['stopwords_removed'] = lyrics_train['no_punc'].apply(lambda x: [word for word in x if word not in stop_words])
lyrics_train.head()
```

```python
import nltk
nltk.download('averaged_perceptron_tagger')

lyrics_train['pos_tags'] = lyrics_train['stopwords_removed'].apply(nltk.tag.pos_tag)
lyrics_train.head()
```

```python
!pip install nltk

def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
    
lyrics_train['wordnet_pos'] = lyrics_train['pos_tags'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])


lyrics_train.head()
```

```python
wnl = WordNetLemmatizer()
lyrics_train['lemmatized'] = lyrics_train['wordnet_pos'].apply(lambda x: [wnl.lemmatize(word, tag) for word, tag in x])
lyrics_train.tail()
```
